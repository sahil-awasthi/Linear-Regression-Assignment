 ## Loading and inspecting the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Load the dataset
df = pd.read_csv('day.csv')

# Display basic information about the dataset
print(df.info())
print(df.describe())
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Check the column names
print(df.columns)
```

## Converting Categorical Variables

```python
# Convert 'season', 'weathersit', 'mnth', 'weekday', 'holiday', 'workingday' into categorical variables
categorical_cols = ['season', 'weathersit', 'mnth', 'weekday', 'holiday', 'workingday']

# Check which of these columns are present in the dataframe
existing_categorical_cols = [col for col in categorical_cols if col in df.columns]

# Convert the existing columns to categorical
df[existing_categorical_cols] = df[existing_categorical_cols].astype('category')

# Create dummy variables
df = pd.get_dummies(df, drop_first=True)
```

## Defining the Target Variable and Dropping Columns

```python
# Define the target variable
y = df['cnt']

# Define the columns to drop
columns_to_drop = ['cnt', 'casual', 'registered', 'dteday', 'instant']

# Drop the columns that exist in the dataframe
X = df.drop([col for col in columns_to_drop if col in df.columns], axis=1)
```

## Train-Test Split and Feature Scaling

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data
X_test_scaled = scaler.transform(X_test)
```

## Building the Model

```python
# Initialize the linear regression model
lr = LinearRegression()

# Fit the model to the training data
lr.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = lr.predict(X_test_scaled)

# Calculate the R-squared score
r2 = r2_score(y_test, y_pred)
print(f'R-squared score on test set: {r2}')
```

# R-squared score on test set: 0.7433217286089682

### Checking for Multicollinearity

```python
# Calculate the correlation matrix
corr_matrix = pd.DataFrame(X_train_scaled, columns=X.columns).corr()

# Display the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Display pairs of highly correlated features
high_corr_pairs = [(i, j) for i in corr_matrix.columns for j in corr_matrix.columns if i != j and abs(corr_matrix.at[i, j]) > 0.8]
print("Highly correlated pairs of features:", high_corr_pairs)
```

### Removing Highly Correlated Features

```python
# Drop one feature from each pair of highly correlated features
features_to_drop = set()
for i, j in high_corr_pairs:
    features_to_drop.add(j)  # You can choose to drop either i or j

X_train_reduced = X_train.drop(columns=features_to_drop)
X_test_reduced = X_test.drop(columns=features_to_drop)

# Scale the reduced features
X_train_scaled_reduced = scaler.fit_transform(X_train_reduced)
X_test_scaled_reduced = scaler.transform(X_test_reduced)

# Calculate VIF for the reduced feature set
vif_data_reduced = pd.DataFrame()
vif_data_reduced['Feature'] = X_train_reduced.columns
vif_data_reduced['VIF'] = [variance_inflation_factor(X_train_scaled_reduced, i) for i in range(X_train_scaled_reduced.shape[1])]
print(vif_data_reduced)
```

### Building the Model with Reduced Features

```python
# Initialize the linear regression model
lr_reduced = LinearRegression()

# Fit the model to the training data with reduced features
lr_reduced.fit(X_train_scaled_reduced, y_train)

# Predict on the test set with reduced features
y_pred_reduced = lr_reduced.predict(X_test_scaled_reduced)

# Calculate the R-squared score for the reduced feature set
r2_reduced = r2_score(y_test, y_pred_reduced)
print(f'R-squared score on test set with reduced features: {r2_reduced}')
```

## Evaluating Multicollinearity using Variance Inflation Factor (VIF)

```python
# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_train_scaled, i) for i in range(X_train_scaled.shape[1])]
print(vif_data)
```

## Residual Analysis

```python
# Residual analysis
residuals = y_test - y_pred

# Plot residuals
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')
plt.show()

# Q-Q plot
import statsmodels.api as sm
sm.qqplot(residuals, line='45')
plt.title('Q-Q Plot')
plt.show()
```

## Exploratory Data Analysis (EDA)

```python
# Pairplot to visualize relationships between numerical variables
sns.pairplot(df[['cnt', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered']])
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

##  Linear Regression Algorithm

```python
# Example data for linear regression illustration
np.random.seed(0)
X_example = np.random.rand(100, 1) * 10
y_example = 2.5 * X_example + np.random.randn(100, 1) * 5

# Fit the model
lr_example = LinearRegression()
lr_example.fit(X_example, y_example)
y_pred_example = lr_example.predict(X_example)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X_example, y_example, color='blue', label='Data Points')
plt.plot(X_example, y_pred_example, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

## Anscombe’s Quartet

```python
import seaborn as sns
import statsmodels.api as sm

# Load Anscombe's quartet dataset
anscombe = sns.load_dataset("anscombe")

# Plot
plt.figure(figsize=(10, 10))
sns.lmplot(x='x', y='y', col='dataset', hue='dataset', data=anscombe, 
           col_wrap=2, ci=None, palette="muted", height=4, scatter_kws={"s": 50, "alpha": 1})
plt.show()
```

## Pearson’s R

```python
# Example data for Pearson's R illustration
np.random.seed(1)
X_pearson = np.random.rand(100)
y_pearson = 2 * X_pearson + np.random.randn(100) * 0.5

# Calculate Pearson's correlation coefficient
pearson_r = np.corrcoef(X_pearson, y_pearson)[0, 1]
print(f"Pearson's R: {pearson_r}")

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X_pearson, y_pearson, color='purple', label=f'Pearson\'s R: {pearson_r:.2f}')
plt.plot(np.unique(X_pearson), np.poly1d(np.polyfit(X_pearson, y_pearson, 1))(np.unique(X_pearson)), color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Pearson Correlation')
plt.legend()
plt.show()
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Example data for scaling illustration
X_scaling = np.random.rand(100, 1) * 100

# Normalize
scaler_norm = MinMaxScaler()
X_normalized = scaler_norm.fit_transform(X_scaling)

# Standardize
scaler_std = StandardScaler()
X_standardized = scaler_std.fit_transform(X_scaling)

# Plot original data
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(X_scaling, bins=20, color='blue', edgecolor='black')
plt.title('Original Data')

# Plot normalized data
plt.subplot(1, 3, 2)
plt.hist(X_normalized, bins=20, color='green', edgecolor='black')
plt.title('Normalized Data')

# Plot standardized data
plt.subplot(1, 3, 3)
plt.hist(X_standardized, bins=20, color='red', edgecolor='black')
plt.title('Standardized Data')

plt.show()
```
